```python
# Define the todo list as an empty list of tasks
todo_list = []

# Function to add a new task
def add_task():
    task = input("Enter the task description: ")
    category = input("Enter the category for the task: ")
    due_date = input("Enter the due date (DD/MM/YYYY): ")
    due_time = input("Enter the due time (HH:MM): ")
    
    task_info = {
        'task': task,
        'category': category,
        'due_date': due_date,
        'due_time': due_time
    }
    
    todo_list.append(task_info)
    print("Task added successfully!")

# Function to edit an existing task
def edit_task():
    list_tasks()  # Show the list of tasks to the user
    task_id = int(input("Enter the task number to edit: ")) - 1
    if 0 <= task_id < len(todo_list):
        task = input("Enter the new task description: ")
        category = input("Enter the new category: ")
        due_date = input("Enter the new due date (DD/MM/YYYY): ")
        due_time = input("Enter the new due time (HH:MM): ")
        
        todo_list[task_id] = {
            'task': task,
            'category': category,
            'due_date': due_date,
            'due_time': due_time
        }
        print("Task updated successfully!")
    else:
        print("Invalid task number.")

# Function to delete a task
def delete_task():
    list_tasks()  # Show the list of tasks to the user
    task_id = int(input("Enter the task number to delete: ")) - 1
    if 0 <= task_id < len(todo_list):
        todo_list.pop(task_id)
        print("Task deleted successfully!")
    else:
        print("Invalid task number.")

# Function to list all tasks
def list_tasks():
    if not todo_list:
        print("No tasks to display.")
    else:
        for idx, task_info in enumerate(todo_list, 1):
            print(f"{idx}. Task: {task_info['task']}, Category: {task_info['category']}, "
                  f"Due Date: {task_info['due_date']}, Due Time: {task_info['due_time']}")

# Function to list tasks by category
def list_by_category():
    category = input("Enter the category to filter by: ")
    filtered_tasks = [task for task in todo_list if task['category'] == category]
    
    if not filtered_tasks:
        print(f"No tasks found in the category '{category}'.")
    else:
        for idx, task_info in enumerate(filtered_tasks, 1):
            print(f"{idx}. Task: {task_info['task']}, Due Date: {task_info['due_date']}, Due Time: {task_info['due_time']}")

# Main function to interact with the user
def main():
    while True:
        print("\nTodo List Options:")
        print("1. Add Task")
        
        # Only show edit task option if there are tasks in the list
        if todo_list:
            print("2. Edit Task")
            print("3. Delete Task")
        print("4. List All Tasks")
        print("5. List Tasks by Category")
        print("6. Exit")
        
        choice = input("Choose an option: ")
        
        if choice == '1':
            add_task()
        elif choice == '2' and todo_list:  # Check if there are tasks to edit
            edit_task()
        elif choice == '3' and todo_list:  # Check if there are tasks to delete
            delete_task()
        elif choice == '4':
            list_tasks()
        elif choice == '5':
            list_by_category()
        elif choice == '6':
            print("Exiting program. Goodbye!")
            break
        else:
            if choice == '2':
                print("No tasks available to edit.")  # Inform user if no tasks are present
            elif choice == '3':
                print("No tasks available to delete.")  # Inform user if no tasks are present
            else:
                print("Invalid option. Please try again.")

# Run the program
if __name__ == "__main__":
    main()


```
